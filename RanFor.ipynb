```python
# .iloc[linhas, colunas] (comando do pandas que faz divisão de dados)
Caracteristicas = Base_Dados.iloc[:, 1:4].values
Previsor = Base_Dados.iloc[:, 4:5].values
```

```python
from sklearn.model_selection import train_test_split

x_treinamento, x_teste, y_treinamento, y_teste = train_test_split(
    Caracteristicas,
    Previsor,
    test_size=0.30, # Divide 70% para treinamento e 30% para teste
    random_state=10
)
print('Linhas p/dados:',len(Base_Dados),'\nLinhas p/treinamento:',len(x_treinamento),'\nLinhas p/teste:',len(x_teste))
```

```python
from sklearn.ensemble import RandomForestClassifier
# Importamos o conjunto de fórmulas da floresta aleatória
Algoritmo_rf = RandomForestClassifier(n_estimators=500)
Algoritmo_rf.fit(x_treinamento, y_treinamento)
# Vai encontrar a melhor árvore para o nosso conjunto de dados
```

```python
Previsoes = Algoritmo_rf.predict(x_teste)
from sklearn.metrics import confusion_matrix
Matriz_Confusao = confusion_matrix(y_teste, Previsoes)
print(Matriz_Confusao)

plt.figure(figsize=(10,5))
sns.heatmap(Matriz_Confusao, annot=True);
```

```python
from sklearn.metrics import classification_report
# Mostrará o percentual de assertividade
report = classification_report(y_teste, Previsoes)
print(report)
```
